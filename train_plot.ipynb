```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline

miniImagenet_1shot_baseline = pd.read_csv('csv/miniImagenet_1shot_baseline.csv', header=None)
miniImagenet_1shot_tcmaml = pd.read_csv('csv/miniImagenet_1shot_tcmaml.csv', header=None)

numrows = miniImagenet_1shot_baseline.shape[0]

miniImagenet_1shot_baseline_uc = miniImagenet_1shot_baseline.iloc[:,-1]
miniImagenet_1shot_tcmaml_uc = miniImagenet_1shot_tcmaml.iloc[:,-1]

f1 = plt.figure(figsize=(5, 5))
#plt.title("mini-ImageNet 5-way 1-shot")
plt.plot(range(numrows), miniImagenet_1shot_baseline_uc, label='MAML', lw=0.5)
plt.plot(range(numrows), miniImagenet_1shot_tcmaml_uc, 'orange', label='TCMAML', lw=0.5)
plt.ylim(3.9,5)
plt.xlabel("Epoch", size=12)
plt.ylabel("Average Task Uncertainty", size=12)
plt.legend(loc=1)
plt.savefig('miniImagenet_uc.png')
plt.savefig('miniImagenet_uc.eps')
```

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline
import os

def train_plot(dataset, shot, baseline_csv, tcmaml_csv, y_min, y_max, legend=True, label='TCMAML'):
    baseline = pd.read_csv(baseline_csv, header=None)
    tcmaml = pd.read_csv(tcmaml_csv, header=None)

    numrows = baseline.shape[0]

    baseline_uc = baseline.iloc[:,-1]
    tcmaml_uc = tcmaml.iloc[:,-1]

    f1 = plt.figure(figsize=(5, 5))

    plt.plot(range(numrows), baseline_uc, label='MAML', lw=0.5)
    plt.plot(range(numrows), tcmaml_uc, 'orange', label=label, lw=0.5)
    plt.ylim(y_min, y_max)
    plt.xlabel("Epoch", size=12)
    plt.ylabel("Average Task Uncertainty", size=12)
    if legend:
        plt.legend(loc=1)
    savepath = os.path.join('save', dataset+str(shot)+'_uc_train')
    plt.savefig(savepath + '.png')
    plt.savefig(savepath + '.eps')
```

```python
train_plot('miniImagenet', 1, 'csv/miniImagenet_1shot_baseline_train.csv', 'csv/miniImagenet_1shot_tcmaml_train.csv', 3.8, 5)
```

```python
train_plot('cross', 1, 'csv/cross_1shot_baseline_train.csv', 'csv/cross_1shot_tcmaml_train.csv', 4.0, 5)
```

```python
train_plot('cross', 5, 'csv/cross_5shot_baseline_train.csv', 'csv/cross_5shot_tcmaml_train.csv', 3.0, 4.8, False)
```
