```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline

#MAML 5-way 5-shot mini-ImageNet
x = np.arange(0.2,1,0.1)
height = np.array([ 0.2594,	0.3473,	0.4304,	0.5199,	0.6133,	0.6892,	0.7878,	0.892])
height2 = np.array([ 0.25, 0.35, 0.45, 0.55, 0.65, 0.75, 0.85, 0.95])
difference = [height[i] - height2[i] if height[i] > height2[i] else 0 for i in range(8)]

fig, ax = plt.subplots(figsize=(5,5))
plt.bar(x, height2, width=0.1, align='edge', color='c', alpha=0.5)
plt.bar(x, height, width=0.1, align='edge', color='r', edgecolor='black')
plt.bar(x, difference, bottom=x+0.05, width=0.1, align='edge', color='c', alpha=0.4)
plt.plot([0,1], [0,1], color='gray', linestyle='dashed')

plt.xlabel("Confidence", size=12)
plt.ylabel("Accuracy", size=12)

textstr = 'ECE = 0.04063 \nMCE = 0.06293'
props = dict(boxstyle='round', facecolor='gray', alpha=0.2)
ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=12,
        verticalalignment='top', bbox=props)

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.savefig('save/maml_mini_5shot.png')
plt.savefig('save/maml_mini_5shot.pdf')
```

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt
import numpy as np
%matplotlib inline

def rdplot(acc_list, textstr, filename):
    x = np.arange(0.2,1,0.1)
    height = np.array(acc_list)
    height2 = np.array([0.25, 0.35, 0.45, 0.55, 0.65, 0.75, 0.85, 0.95])
    difference = [height[i] - height2[i] if height[i] > height2[i] else 0 for i in range(8)]

    fig, ax = plt.subplots(figsize=(5,5))
    plt.bar(x, height2, width=0.1, align='edge', color='c', alpha=0.5)
    plt.bar(x, height, width=0.1, align='edge', color='r', edgecolor='black')
    plt.bar(x, difference, bottom=x+0.05, width=0.1, align='edge', color='c', alpha=0.4)
    plt.plot([0,1], [0,1], color='gray', linestyle='dashed')

    plt.xlabel("Confidence", size=12)
    plt.ylabel("Accuracy", size=12)

    textstr = textstr
    props = dict(boxstyle='round', facecolor='gray', alpha=0.2)
    ax.text(0.05, 0.95, textstr, transform=ax.transAxes, fontsize=12,
            verticalalignment='top', bbox=props)

    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.savefig(filename+'.png')
    plt.savefig(filename+'.pdf')
```
