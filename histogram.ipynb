```python
import matplotlib.pyplot as plt
import numpy as np
import os
%matplotlib inline

def hist(dataset, shot, epoch, min, max, bins, legend=True):
    baseline_path = os.path.join('save/training_uc', dataset+str(shot)+'_'+str(epoch)+'epoch_baseline.npy')
    tcmaml_path = os.path.join('save/training_uc', dataset+str(shot)+'_'+str(epoch)+'epoch_tcmaml.npy')
    baseline_uncertainty = np.load(baseline_path)
    tcmaml_uncertainty = np.load(tcmaml_path)

    fig, ax = plt.subplots(figsize=(5,2))
    plt.hist(baseline_uncertainty, bins=bins, range=(min, max), alpha=0.7, rwidth=0.8, label='MAML')
    plt.hist(tcmaml_uncertainty, bins=bins, range=(min, max),  alpha=0.7, rwidth=0.8, label='TCMAML', color='orange')
    if legend:
        plt.legend(loc=2)
    plt.xlabel("Task Uncertainty "+str(epoch)+" epoch", size=10)
    filename = 'save/training_uc/histogram/' + dataset+str(shot)+'_'+str(epoch)+'epoch.pdf'
    plt.savefig(filename,bbox_inches='tight')
```
