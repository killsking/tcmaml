```python
# pretrained on 5-way 5-shot mini-ImageNet
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

task_uncertainty_all = np.load('save/task_uc_mini.npy')
acc_all = np.load('save/task_acc_mini.npy')
task_uncertainty_all_cub = np.load('save/task_uc_cub.npy')
acc_all_cub = np.load('save/task_acc_cub.npy')
label =  ['mini-ImageNet']*600 + ['CUB-200']*600 

df = pd.DataFrame()
# Add columns
df['x'] = np.concatenate((task_uncertainty_all,task_uncertainty_all_cub))
df['y'] = np.concatenate((acc_all,acc_all_cub))
df['Testset'] = label
sns.set()
fig, ax = plt.subplots(figsize=(8,6))

#df1 = pd.DataFrame()
# Add columns
#df1['x'] = task_uncertainty_all_cub
#df1['y'] = acc_all_cub
#sns.set()

#sns.scatterplot(x="x", y="y", data=df1, ax=ax, s=30, palette='colorblind')
colors = ['#143b6e', '#931717']
sns.set_palette(sns.color_palette(colors))
sns.scatterplot(x="x", y="y", data=df, ax=ax, s=30, hue='Testset')
ax.set_xlabel('Task Uncertainty')
ax.set_ylabel('Task Accuracy (%)')
fig.show()
fig.savefig('tuc.eps')
fig.savefig('tuc.png')
```
